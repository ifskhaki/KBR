## âš™ï¸ Configuration


```python
# Upload address.txt
from google.colab import files
uploaded = files.upload()
if 'address.txt' in uploaded:
    print("âœ… address.txt uploaded")
else:
    print("âš ï¸ Creating example file...")
    !echo "1A1zP1eP5QGefi2DMPTfTL5SLmv7DivfNa" > address.txt
```

```python
# Configure (optional email setup)
import json
config = {
    "email": {"enabled": False},
    "gpu": {"use_all_gpus": True, "auto_optimize": True},
    "search": {"targets_file": "address.txt", "output_file": "Success.txt", 
               "compression": "UNCOMPRESSED", "random256": True},
    "display": {"real_time": True, "update_interval_ms": 1000}
}
!mkdir -p config
with open('config/config.json', 'w') as f:
    json.dump(config, f, indent=4)
print("âœ… Configuration saved")
```
